```python
import sqlite3
import msgpack
from datetime import datetime


conn = sqlite3.connect("/code/my-chat-app/backend/chathistory/sqlite/chathistory.db", check_same_thread=False)
cursor = conn.cursor()

thread_id = "user_session_1"

cursor.execute("SELECT checkpoint_id, checkpoint FROM checkpoints WHERE thread_id = ? ", (thread_id,))
results = cursor.fetchall()

for i, result in enumerate(results):
    checkpoint_id = result[0]  # checkpoint_id
    checkpoint_bytes = result[1]  # checkpoint データ

    
    # checkpoint_bytes が文字列の場合はエンコードする
    if isinstance(checkpoint_bytes, str):
        checkpoint_bytes = checkpoint_bytes.encode('utf-8')
    
    checkpoint_data = msgpack.unpackb(checkpoint_bytes, raw=False)
    print(f"Checkpoint {i}: {checkpoint_id}")
    print(checkpoint_data)
    print("---")

```
